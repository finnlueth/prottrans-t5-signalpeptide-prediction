```python
import seaborn as sns
```

```python
sns.set_theme(style="darkgrid")
fmri = sns.load_dataset("fmri")
my_plot = sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
fig = my_plot.get_figure()
fig.savefig("./plots/out.png") 
```
